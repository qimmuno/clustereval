# Comparing generalized homogeneity and parsimony scores

This notebook compares the information-theoretic, set-matching, and pair-counting homogeneity and parsimony scores on a simple example problem.

```python
import numpy as np
from clustereval import *
```

```python
class_partition = [0, 0, 0, 1, 1, 2, 2, 2]
cluster_partition = [3, 3, 3, 3, 4, 4, 4, 4]
```

## Information-theoretic validation (Shannon entropy)

```python
entropies = conditional_entropies(class_partition, cluster_partition)
entropies.class_given_cluster_entropy/np.log(2), entropies.cluster_given_class_entropy/np.log(2)
```

```python
homogeneity_score(class_partition, cluster_partition), parsimony_score(class_partition, cluster_partition)
```

## Set-matching validation (min entropy)

```python
purity_score(class_partition, cluster_partition), inverse_purity_score(class_partition, cluster_partition)
```

```python
normalized_purity_score(class_partition, cluster_partition), normalized_inverse_purity_score(class_partition, cluster_partition)
```

## Pair-counting validation (collision entropy)

```python
pair_specificity_score(class_partition, cluster_partition), pair_sensitivity_score(class_partition, cluster_partition)
```
